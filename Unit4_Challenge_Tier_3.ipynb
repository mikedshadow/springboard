**2.2. Cleaning the data**

The first step will be eliminating the extraneous columns in the list above.
The first column 'unnamed: 0' holds the dates so we keep it. The next 33 columns contain the average prices by month in the desired boroughs, they are also kept. Unnamed 34,37 and 47 are all empty so they should be dropped. All the rest after 'Westminster' are not the boroughs but other subdivisions of London, so they should also be discarded.
The 0th row is just a code for the borough which is also extraneous

```python
# Use subsetting with iloc to remove the columns
properties = properties.iloc[ : ,:34]
# drop the 0th row
properties = properties.drop(0)
properties.head()
```

This looks much better! I suspect that "City of London' is the aggregate that month from all boroughs but that is another issue. Firstly, the different boroughs should be the rows and the different dates should be columns with entries being average prices in borough(row) on date(column). Even better would be a tall table with borough, month and average price as columns. This can be done by transposing and then melting.

First transpose.

```python
# turns out City of London is not a borough so remove it
properties = properties.drop('City of London', axis=1)
properties.head()
```

```python
properties_T = properties.transpose()
properties_T.shape
```

```python
properties_T.head()
```

```python
# reset the index so the boroughs are not the index
properties_T = properties_T.reset_index()
# set the 0th row as the column headers
properties_T.columns = properties_T.iloc[0]
# drop the now extra 0th row
properties_T = properties_T.drop(0)
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
properties_T = properties_T.rename(columns = {'Unnamed: 0':'London_Borough'})
properties_T.head()
```

**2.4.Transforming the data**

Looks ready to **melt** 
 

```python
clean_properties = pd.melt(properties_T, id_vars= ['London_Borough'])
clean_properties.shape
```

Rename columns to 'month'  and 'Average_price'. Check column data types.

```python
clean_properties = clean_properties.rename(columns = {0: 'Month', 'value': 'Average_price'})
clean_properties.dtypes
clean_properties.shape
```

```python
clean_properties['Average_price'] = pd.to_numeric(clean_properties['Average_price'])
clean_properties.dtypes
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Rename cleaned dataframe as df
df = clean_properties
# peek at line graphs by borough
for b in df['London_Borough'].unique():
    borough_prices = df[df['London_Borough'] == b]
    visual = borough_prices.plot(kind ='line', x = 'Month', y='Average_price') 
    visual.set_ylabel('Price')
    visual.set_title(b)
    plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d, year):
    """This function returns the price ratio in the average price column of d between year and 20 years earlier"""
    year_20base = float(d['Average_price'][d['Year']== year - 20])
    year_avg = float(d['Average_price'][d['Year']== year])
    ratio = [year_avg/year_20base]
    return ratio
```

```python
# test  the function

create_price_ratio(df_by_year[df_by_year['London_Borough']=='Barking & Dagenham'], 2018)
```

```python
# create blank dataframe for ratios
ratios = {}
years = list(range(2015,2023))
for b in df_by_year['London_Borough'].unique():
    # Let's make our parameter to our create_price_ratio function: i.e., we subset dfg on 'London_Borough' == b. 
    borough = df_by_year[df_by_year['London_Borough'] == b ]
    column = []
    for year  in years:
        # Make a new column whose value's the result of calling create_price_ratio with the argument: borough, year
        column.append(create_price_ratio(borough, year))
    ratios[b] = column
df_ratio = pd.DataFrame(ratios)
df_ratio = df_ratio.T
df_ratio = df_ratio.reset_index()
df_ratio.rename(columns={'index':'Borough', 0: years[0], 1: years[1],2: years[2],3:years[3],4:years[4],5:years[5],6: years[6],7:years[7]}, inplace=True)
df_ratio.head()
```

```python
df_ratio.columns
```

```python
order_2018 = df_ratio[['Borough', 2018]].sort_values(by=2018,ascending=False)
order_2022 = df_ratio[['Borough', 2022]].sort_values(by=2022,ascending=False)
```

```python
# add the order  of boroughs in 2022 to order_2018
order_2018.reset_index()
order_2018['2022'] = list(order_2022['Borough'])
order_2018.head(10)
```
